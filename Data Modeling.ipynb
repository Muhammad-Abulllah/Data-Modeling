```python
try:
    cur.execute("DROP TABLE IF EXISTS timeperiod CASCADE")
    cur.execute("DROP TABLE IF EXISTS experience CASCADE")
    cur.execute("DROP TABLE IF EXISTS Job CASCADE")
    cur.execute("DROP TABLE IF EXISTS Information CASCADE")

    cur.execute("""
        CREATE TABLE Information (
            First_Name VARCHAR(200) NOT NULL,
            Last_Name VARCHAR(200) NOT NULL,
            Age VARCHAR(200) NOT NULL,
            Salary INT PRIMARY KEY
        )
    """)

    cur.execute("""
        CREATE TABLE timeperiod (
            leaves_used VARCHAR(50) NOT NULL,
            leaves_remaining VARCHAR(50) NOT NULL,
            DOJ VARCHAR(50) NOT NULL,
            Salary INT REFERENCES Information(Salary)
        )
    """)

    cur.execute("""
        CREATE TABLE experience (
            Past_exp VARCHAR(50),
            Ratings VARCHAR(50),
            Salary INT REFERENCES Information(Salary)
        )
    """)

    cur.execute("""
        CREATE TABLE Job (
            Designation VARCHAR(50) NOT NULL,
            Unit VARCHAR(50) NOT NULL,
            Salary INT REFERENCES Information(Salary)
        )
    """)

    print("Tables created successfully.")

except psycopg2.Error as e:
    print("Error creating tables:", e)
```

```python

def import_data_from_csv(table_name, csv_file, cur):
    try:
        df = pd.read_csv(csv_file)
        df_cols = df.columns.tolist()
        cols = ','.join(df_cols)
        if 'Unnamed: 0' in df_cols:
            df = df.drop(columns=['Unnamed: 0'])
        for _, row in df.iterrows():
            try:
                cur.execute(
                    f"INSERT INTO {table_name} ({cols}) VALUES ({', '.join(['%s' for _ in df_cols])})",
                    tuple(row)
                )
            except psycopg2.Error as e:
                print(f"Error inserting data into {table_name}: {e}")
    
        print(f"Data imported into {table_name} successfully.")

    except FileNotFoundError:
        print(f"Error: File {csv_file} not found.")
    except Exception as e:
        print(f"Error importing data into {table_name}: {e}")


# Define the order of table imports
import_order = [
    "experience",
   "Information",
    "timeperiod",
    
    "Job"
]

# Import data into each table in the specified order
try:
    for table_name in import_order:
        if table_name in csv_files:
            import_data_from_csv(table_name, csv_files[table_name], cur)

    print("All data imported successfully!")

except psycopg2.Error as e:
    print(f"Error inserting data: {e}")
```
